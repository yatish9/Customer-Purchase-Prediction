```python
final_data.rename(columns = {'InvoiceDate':'Date'}, inplace = True)
final_data.head()
```

```python
clv_freq = '3M'
```

```python
data_group=final_data.reset_index().groupby([
            'Customer ID',
            pd.Grouper(key='InvoiceDate', freq=clv_freq)
            ]).agg({'Date': lambda date: (date.max() - date.min()).days,
                                        'Invoice': lambda num: len(num),
                                        'Quantity': lambda quant: quant.sum(),
                                        'Amountspent': lambda pr: pr.sum()})
```

```python
data_group.head()
```

```python
data_group.columns=['no_of_days','no_of_transactions','no_of_units','Amount_spent']
data_group.head()
```

```python
data_group['avg_order_value']=data_group['Amount_spent']/data_group['no_of_transactions']
data_group.head()
```

```python
purchase_frequency=sum(data_group['no_of_transactions'])/data_group.shape[0]
```

```python
repeat_rate=data_group[data_group.no_of_transactions > 1].shape[0]/data_group.shape[0]
churn_rate=1-repeat_rate
purchase_frequency,repeat_rate,churn_rate
```

```python
data_group['profit_margin']=data_group['Amount_spent']*0.05
```

```python
data_group.head()
```

```python
data_group['cust_lifetime']=(data_group['avg_order_value']*purchase_frequency)/churn_rate
data_group['CLTV']=data_group['cust_lifetime']*data_group['profit_margin']
```

```python
data_group.head()
```

```python
data_group = data_group.reset_index()
data_group.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
target_var = 'Target_Chunk' 
all_features = [x for x in ready_data.columns if x not in ['Customer ID', target_var]]
```

```python
all_features
```

```python
ready_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                    ready_data[all_features], 
                                    ready_data[target_var], 
                                    random_state=0,
                                    test_size=0.25)
```

```python
lin_reg = LinearRegression(n_jobs=1)  
```

```python
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.intercept_
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
coef = pd.DataFrame(list(zip(all_features, lin_reg.coef_)))
coef.columns = ['feature', 'coef']
```

```python
coef
```

```python
from sklearn.metrics import r2_score, median_absolute_error
```

```python
train_preds =  lin_reg.predict(X_train)
test_preds = lin_reg.predict(X_test)
```

```python
print('R-Squared for Train set: %0.2f' % r2_score(y_true=y_train, y_pred=train_preds))
print('R-Squared for Test set: %0.2f' % r2_score(y_true=y_test, y_pred=test_preds))
```

```python
print('R-Squared for Train set: %0.2f' % r2_score(y_true=y_train, y_pred=train_preds))
```

```python
print('MedAE for Train set: %0.2f' % median_absolute_error(y_true=y_train, y_pred=train_preds))
print('MedAE for Test set: %0.2f' % median_absolute_error(y_true=y_test, y_pred=test_preds))
```
